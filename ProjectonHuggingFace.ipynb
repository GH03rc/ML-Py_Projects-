```python
#step 2 Pipeline
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
print(classifier("I love machine learning,gen ai,python,hugging face"))
```

```python
#step 3
def classify_tweet(tweet):
    result = classifier(tweet)[0]
    label = result['label']
    score = round(result['score']*100,5)
    return f"{label}({score}%)"
```

```python
import gradio as gr
```

```python
interface=gr.Interface(
    fn=classify_tweet,
    inputs=gr.Textbox(lines=4,placeholder="Paste Your Tweet Here.... "),
    outputs='text',
    title="Smart Tweet Sentiment Classifier",
    description = "Analyze the sentiment of a tweet using a pretrained hugging face model",
)
interface.launch()
```
